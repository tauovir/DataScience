# Geometric Transformations of Images:

In this section we will learn to apply different geometric transformation to images like translation, rotation, affine transformation etc.

# Transformations:

OpenCV provides two transformation functions, cv2.warpAffine and cv2.warpPerspective, with which you can have all kinds of transformations. cv2.warpAffine takes a 2x3 transformation matrix while cv2.warpPerspective takes a 3x3 transformation matrix as input
# Scaling:
Scaling is just resizing of the image. OpenCV comes with a function cv2.resize() for this purpose. The size of the image can be specified manually, or you can specify the scaling factor. Different interpolation methods are used. Preferable interpolation methods are cv2.INTER_AREA for shrinking and cv2.INTER_CUBIC (slow) & cv2.INTER_LINEAR for zooming. By default, interpolation method used is cv2.INTER_LINEAR for all resizing purposes. You can resize an input image either of following methods:

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python

def displayImage(imageList):
    plt.figure(figsize=(40,20))
    i = 0
    for item in imageList:  
        plt.subplot(2, 5, i+1)
        plt.imshow(item['image'])
        i += 1
        plt.axis('off')
        plt.title(item['name'],fontsize = 18, color = 'green')
    plt.tight_layout()
    plt.show()
    
```

```python
img = cv2.imread('images/opencv.jpg')
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#res = cv2.resize(image_RGB,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)

#OR

height, width = image_RGB.shape[:2]
res = cv2.resize(image_RGB,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
print("Original Height:{height},Width:{width}".format(height = height,width= width))
height1, width1 = res.shape[:2]
print("Extended Height:{height},Width:{width}".format(height = height1,width= width1))
displayImage([
    {'name':'Origina Image','image':image_RGB},
     {'name':'scalled Image','image':res},
             ])
```

# Translation:
Translation is the shifting of object’s location. In order to translate an object, we need to create the 2 x 3 transformation matrix by using the NumPy array with float values providing the translation in both the x and y directions in pixels, as shown in the following code:
M = np.float32(  1, 0, x ,  0, 1, y  )

This gives the following M transformation matrix:
M = [1 0 tx]
    [0 1 ty]

Once this matrix has been created, the cv2.warpAffine() function is called, as shown in the following code
dst image = cv2.warpAffine(image, M, (width, height))

The cv2.warpAffine() function transforms the source image using the M matrix provided. The third (width, height) argument establishes the size of the output image.

For example, if we want to translate an image with 550 pixels in the x direction and 100 pixels in the y direction, we use the following.
NOte: we can also assign negative value for opposit direction.


```python
img = cv2.imread('images/two-birds.jpg')
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

height, width = image_RGB.shape[:2]
M = np.float32([[1,0,550],[0,1,100]])
dst = cv2.warpAffine(image_RGB,M,(width,height))
displayImage([
    {'name':'Origina Image','image':image_RGB},
     {'name':'Translated','image':dst},
             ])
```

# Rotation:
In order to rotate the image, we make use of the cv.getRotationMatrix2D() function to build the 2 x 3 transformation matrix. This matrix rotates the image at the desired angle (in degrees), where positive values indicate a counterclockwise rotation. Both the center of rotation and the scale factor can also be adjusted. Using these elements in our example, the following transformation matrix is calculated

```python
img = cv2.imread('images/two-birds.jpg')
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows,cols = image_RGB.shape[:2]

M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
dst = cv2.warpAffine(image_RGB,M,(cols,rows))
displayImage([
    {'name':'Origina Image','image':image_RGB},
     {'name':'Rotated Image','image':dst},
             ])
```

# Affine Transformation
In affine transformation, all parallel lines in the original image will still be parallel in the output image. To find the transformation matrix, we need three points from input image and their corresponding locations in output image. Then cv2.getAffineTransform will create a 2x3 matrix which is to be passed to cv2.warpAffine.

Note : An affine transformation is a transformation where points, straight lines, and planes are preserved. Additionally, the parallel lines will remain parallel after this transformation. However, an affine transformation does not preserve both the distance and angles between points.

Check below example

```python

img = cv2.imread('images/hand-written.png')
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows,cols,ch = image_RGB.shape

pts1 = np.float32([[135,45],[385,45],[135,230]])
pts2 = np.float32([[135,45],[385,45],[150,230]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(image_RGB,M,(cols,rows))

displayImage([
    {'name':'Origina Image','image':image_RGB},
     {'name':'Affine Transform','image':dst},
             ])
```

# Perspective Transformation
In order to correct the perspective (also known as perspective transformation), we will need to create the transformation matrix by making use of the cv2.getPerspectiveTransform() function, where a 3 x 3 matrix is constructed. This function needs four pairs of points (coordinates of a quadrangle in both the source and output image) and calculates a perspective transformation matrix from these points. Then, the M matrix is passed to cv2.warpPerspective(), where the source image is transformed by applying the specified matrix with a specified size, as shown in the following code:

```python


img = cv2.imread('images/borderPage.png')
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows,cols,ch = image_RGB.shape

pts1 = np.float32([[0,0],[800,10],[0,387],[500,600]])
pts2 = np.float32([[0,0],[300,0],[0,387],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(image_RGB,M,(300,300))

displayImage([
    {'name':'Origina Image','image':image_RGB},
     {'name':'Affine Transform','image':dst},
             ])
```
