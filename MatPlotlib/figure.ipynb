```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.arange(0, math.pi * 2, 0.05)
y = np.sin(x)
x.shape
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  #[left, bottom, width, height]
ax.plot(x,y,'g--')
ax.set_title("Learning Graph")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```
