```python
x = np.arange(0, math.pi * 2, 0.05)
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.60,0.55,0.3,0.3])
y = np.sin(x)
ax1.plot(x,y,'b')
ax2.plot(x, np.cos(x),'g')
ax1.set_title('Sine')
ax1.set_xlabel('animal')
ax1.set_ylabel('Birds')

ax2.set_title("cosine")
ax2.set_xlabel('Mango')
ax2.set_ylabel('Apple')
plt.show()
```

```python

#Plt.subplots(nrows, ncols)
#he function returns a figure object and a tuple containing axes objects equal to nrows*ncols. fig,tubple = plt.subplots(2,3)
fig, tuple1 = plt.subplots(2,2)
x = np.arange(1,5)
tuple1[0][0].plot(x, x * x)
tuple1[0][0].set_title("Square")
tuple1[0][1].plot(x,np.sqrt(x))
tuple1[0][1].set_title('square root')
tuple1[1][0].plot(x,np.exp(x))
tuple1[1][0].set_title('exp')
tuple1[1][1].plot(x,np.log10(x))
tuple1[1][1].set_title('log')
plt.show()
```
