```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

def displayImage(imageList,cmap=None):
    plt.figure(figsize=(18,12))
    i = 0
    for item in imageList:  
        plt.subplot(2, 5, i+1)
        plt.imshow(item['image'],cmap)
        i += 1
        plt.axis('off')
        plt.title(item['name'],fontsize = 18, color = 'green')
    plt.tight_layout()
    plt.show()
    
```
