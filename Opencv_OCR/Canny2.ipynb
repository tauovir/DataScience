```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

def displayImage(imageList,cmap=None):
    plt.figure(figsize=(18,12))
    i = 0
    for item in imageList:  
        plt.subplot(2, 5, i+1)
        plt.imshow(item['image'],cmap)
        i += 1
        plt.axis('off')
        plt.title(item['name'],fontsize = 18, color = 'green')
    plt.tight_layout()
    plt.show()
    
```

```python

img = cv2.imread('images/opencv.jpg') 
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,cv2.THRESH_BINARY)

imgBlur = cv2.GaussianBlur(imgray,(5,5),1)
#ret,thresh1 = cv2.threshold(imgBlur,150,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C)
Canny = cv2.Canny(imgBlur,80,230)
contours, heirarchy = cv2.findContours(Canny,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img, contours, -1, (0,250,0), 3)
print("Conours:{}".format(str(len(contours))))
displayImage([
        {'name':'Original Image1','image':img},
        {'name':'Gray Image','image':imgray},
        {'name':'Gray Image','image':Canny},
        ],'gray')
```
