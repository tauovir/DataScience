## Find this Error : invalid literal for int() with base 10: 'Or'

```python
err = dataset[dataset['Order Date'].str[0:2] == 'Or']
err.head()
```

```python
## get rid of above data
dataset =  dataset[dataset['Order Date'].str[0:2] != 'Or']
```

## Lets see what was the best month for Data? How much we earned for that month

## First Add Month Columns

```python
dataset.head()
```

```python
dataset['month'] = dataset['Order Date'].str[0:2]
## Conver it to int
dataset['month'] = dataset['month'].astype('int32')
```

```python
dataset.head()
```

## Now Add Sales Data

## First Convert Quantity Ordered and Price Each to numeric type;
We can use astype, but we are using pd.to_numeric beacuse pandas will set correct format

```python
dataset['Quantity Ordered'] = pd.to_numeric(dataset['Quantity Ordered']) ## Pandas will set correct numeric type
dataset['Price Each'] = pd.to_numeric(dataset['Price Each']) 
```

```python
dataset['sales'] = dataset['Quantity Ordered'] * dataset['Price Each'].astype('float')
dataset.head()
```

```python
# No we can find our first question
result = dataset.groupby('month').sum()
# dataset.groupby('month').sum()['sales'] #To check only sales data
result
```

## Show Sales in Bar Graph

```python
import datetime
def monthNumToName():
    monthList = []
    months =  range(1,13)
    for mn in months:
        month = datetime.date(1900, mn, 1).strftime('%b') # B full name
        monthList.append(month)
    return monthList
```

```python
months =  monthNumToName()
plt.bar(months, result['sales'])
plt.xlabel("Month Number")
plt.ylabel("Sales in US($)")
plt.title("Sales Data Analysis")
plt.xticks(months)
plt.show()
```

## Add City Columns
Split Purchase address to extract City


```python
def getCity(address):
    return address.split(",")[1]
    
```

```python

# dataset['City'] = dataset['Purchase Address'].apply(lambda x : x.split(",")[1] ) # It will apply in all column
dataset['City'] = dataset['Purchase Address'].apply(lambda x : getCity(x) ) # It will apply in all column
dataset.head(10)
```

```python

plt.bar(cityWiseSale.index, cityWiseSale['sales'])
plt.xlabel("City Name")
plt.ylabel("Sales in US($)")
plt.title("Sales Data Analysis")
plt.xticks(cityWiseSale.index,fontsize=10, rotation=90)
plt.show()
```

## What product most often sold together

```python
dataset.head(10)
```

```python
duplicateOrder = dataset[dataset['Order ID'].duplicated(keep = False)]
duplicateOrder.head()
```

```python
duplicateOrder['grouped'] = duplicateOrder.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
duplicateOrder.head(10)
```

```python
duplicateOrder = duplicateOrder[['Order ID', 'grouped']].drop_duplicates()
duplicateOrder.head()
```

## Count the product to find most occurence

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in duplicateOrder['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2))) # 2 product combination, or we can cahnge 3 combination
    
```

```python
#print(count.most_common(10))
for k,v in count.most_common(10):
    print(k,v)
```

## What product Sold the Most

```python
product_group = dataset.groupby('Product')
quantityOrder = product_group.sum()['Quantity Ordered']
```

```python
quantityOrder
```

```python
plt.bar(quantityOrder.index, quantityOrder)
plt.xlabel("Product Name")
plt.ylabel("Product Quantity")
plt.title("Sales Data Analysis")
plt.xticks(quantityOrder.index,fontsize=10, rotation=90)
plt.show()
```

```python
prices = dataset.groupby('Product').mean()['Price Each']
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(quantityOrder.index, quantityOrder, color = 'g')
ax2.plot(quantityOrder.index, prices, '-b')
ax1.set_xlabel("Product Name")
ax1.set_ylabel("Product Quantity")
ax2.set_ylabel("Price $", color = 'b')

plt.title("Sales Data Analysis")
ax1.set_xticklabels(quantityOrder.index,fontsize=10, rotation=90)
plt.show()
```
