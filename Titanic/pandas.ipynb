# Data Cleaning

```python
data.head(2)
```

```python
data.isnull().head(3) # Checks for null Values, Returns Boolean Arrray
```

```python
data.notnull().head(3) # Opposite of pd.isnull()
```

```python
data1 = data.copy()
data1.shape
```

```python
data1.dropna(inplace = True) # Drop all rows that contain null values
```

```python
data1.shape
```

```python
data2 = data.copy()
data2.shape
```

```python
data2.dropna(axis=1, inplace = True) # Drop all columns that contain null values
data2.shape
```

```python
data3 = data.copy()
data3['Age'].fillna(data3['Age'].mean(),inplace = True) # s.fillna(s.mean()), Replace all null values with the mean
```

```python
data3['Age'].isnull().sum()
```

```python
#s.astype(float)
data['Age'] = data['Age'].astype(float)
```

```python
# s.replace(1,'one'),s.replace([2,3],['two', 'three'])
data['Sex'].replace('male-Per', 'male-Per55',inplace = True)
data['Sex'].head(5)
```

```python
#df.rename(columns={'old_name': 'new_ name'}) # Selective renaming
data.rename(columns={'PassengerId': 'pass_id'},inplace = True)
data.head(5)
```

# Filter, Sort, and Groupby

```python
data[data['Age'] >= 30].head(3) # Rows where the column col is greater than= 30
```

```python
data[(data['Age'] >= 30) & (data['Age'] < 35)].head(3) #Rows where 35 > col > 30
```

```python
# df.sort_values(col1) # Sort values by col1 in ascending order
data.sort_values('Age',ascending = False).head(5)
```

```python
# df.sort_values([col1,col2],ascending=[True,False]) # Sort values by col1 in ascending order then col2 in descending order
data.sort_values(['Pclass','Age'],ascending=[False,True]).head(6)
```
