```python
# Common Operator used in Dictionary
class OperatorIDict():
    def __init__(self,dict1):
        self.dict1 = dict1
        print(self.dict1)
    
    """
    the in and not in operators return True or False according to whether 
    the specified operand occurs as a key in the dictionary:

    """
    def in_operator1(self,searchKey):
        print("Length:{length}".format(length = len(self.dict1)))
        if searchKey in self.dict1:
            print("{searchKey} exist in Dictionary".format(searchKey = searchKey))
        else:
            print("{searchKey} not exist in Dictionary".format(searchKey = searchKey))
            
    def in_operator2(self,searchKey):
        
        if searchKey not in self.dict1:
            print("{searchKey} not exist in Dictionary".format(searchKey = searchKey))
        else:
            print("{searchKey} exist in Dictionary".format(searchKey = searchKey))
            
        
        #You can use the in operator together with short-circuit evaluation to avoid raising
        #an error when trying to access a key that is not in the dictionary:
            
    def in_operator3(self,searchKey):
        
        if searchKey in self.dict1 and self.dict1[searchKey]:
            print("{searchKey} exist in Dictionary".format(searchKey = searchKey))
        else:
            print("{searchKey} not exist in Dictionary".format(searchKey = searchKey))
            
       
dict1 = dict(
    name = 'Taukir khan',
    city = 'Meerut',
    designation = 'Developer',
    qualification = 'Master',
    isWorking = True
)
obj1 = OperatorIDict(dict1)
obj1.in_operator1('city')

    
    
        
```

# Dictionary In built Function

```python
"""
It allows you to create a set of methods that must be created within any child classes built from the abstract class. 
A class which contains one or more abstract methods is called an abstract class. An abstract method is a method 
that has a declaration but does not have an implementation.

How Abstract Base classes work :
By default, Python does not provide abstract classes. Python comes with a module which provides the base for defining 
Abstract Base classes(ABC) and that module name is ABC. ABC works by decorating methods of the base class as abstract and 
then registering concrete classes as implementations of the abstract base. 
A method becomes abstract when decorated with the keyword @abstractmethod. 
"""
from abc import ABC, abstractmethod 
class DictBuiltFunc(ABC):
    #empties dictionary of all key-value pairs:
    def clearDict(self):
        pass
    def getKeyDict(self):
        pass
    
    def dictItems(self):
        pass
    
    def dictKeys(self):
        pass
    
    def dictValues(self):
        pass
    
    def dictPop(self):
        pass
    
    def dictPopItem(self):
        pass
    def dictUpdate(self):
        pass
    
class ImplementDict(DictBuiltFunc):
    def __init__(self, dict1):
        self.dict1 = dict1
        print("Original Dictionary")
        print(self.dict1)
    # overriding abstract method 
    def clearDict(self):
        self.dict1.clear()
        print(self.dict1)
        
    def dictItems(self):
        list1 = self.dict1.items() # Returns a list of key-value pairs in a dictionary.
        print(list1)
        
    def dictKeys(self):
        list1 = self.dict1.keys() # Returns a list of keys in a dictionary.
        print("Keys:",list1)
        
    def dictValues(self):
        list1 = self.dict1.values() # Returns a list of values in a dictionary.
        print("Keys:",list1)
        
    def dictPop(self,key):
        # Removes a key from a dictionary, if it is present, and returns its value else reurn default value
        val = self.dict1.pop(key,'Default vale') 
        print("Popped value:",val)
    
    def dictPopItem(self):
        # removes the last key-value pair added from d and returns it as a tuple:
        tuple1 = self.dict1.popitem() 
        print("Popped key-value:",tuple1)
        tuple1 = self.dict1.popitem() 
        print("Popped key-value:",tuple1)
        
    def dictUpdate(self):
        #Merges a dictionary with another dictionary or with an iterable of key-value pairs.
        dict2 = {'gender':'Male','language':'English'}
        self.dict1.update(dict2)
        print("Updated Dictionary:")
        print(self.dict1)

dict1 = dict(
name = 'Taukir khan',
city = 'Meerut',
designation = 'Developer',
qualification = 'Master',
isWorking = True
)
obj1 = ImplementDict(dict1)
# obj1.clearDict()
# obj1.getKeyDict('city2')
# obj1.dictItems()
# obj1.dictKeys()
# obj1.dictValues()
# obj1.dictPop('designation')
# obj1.dictPopItem()
    
obj1.dictUpdate()
```

# OrderedDict in Python:
An OrderedDict is a dictionary subclass that remembers the order that keys were first inserted. The only difference between dict() and OrderedDict() is that:
OrderedDict preserves the order in which the keys are inserted. A regular dict doesn’t track the insertion order, and iterating it gives the values in an arbitrary order. By contrast, the order the items are inserted is remembered by OrderedDict.

```python
from collections import OrderedDict
class OrdDict:
    def __init__(self):
        pass
    def regularDict(self):
        print("This is a Dict:\n") 
        d = {} 
        d['a'] = 1
        d['b'] = 2
        d['c'] = 3
        d['d'] = 4
        for key, value in d.items(): 
            print(key, value) 
            
    def orderDict1(self):
        od = OrderedDict() 
        od['a'] = 1
        od['b'] = 2
        od['c'] = 3
        od['d'] = 4
        print("Order Dict")
        for key, value in od.items(): 
            print(key, value) 
        
obj1 = OrdDict()
obj1.regularDict()
obj1.orderDict1()
```
